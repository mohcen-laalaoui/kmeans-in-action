## **Customer Segmentation Using K-Means Clustering**

### Import Libraries

```python
import numpy as np
import pandas as pd 
import torch
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

### Creating a synthetic dataset

```python
data = {
    'Customer ID': np.arange(1, 201),
    'Age': np.random.randint(18, 70, 200),
    'Annual Income (k$)': np.random.randint(15, 150, 200),
    'Spending Score (1-100)': np.random.randint(1, 100, 200)
}
```

```python
df = pd.DataFrame(data)
```

```python
X = torch.tensor(np.array(list(zip(data['Age'], data['Annual Income (k$)'], data['Spending Score (1-100)']))), dtype=torch.float32)
```

```python
if isinstance(X, np.ndarray):
    X = torch.from_numpy(X).float()  
elif isinstance(X, torch.Tensor):
    X = X.clone().detach().float() 

print("X Shape:", X.shape)
```

### Generate synthetic 2D data

```python
torch.manual_seed(42)
```

```python
scaler = StandardScaler()
X_scaled = torch.tensor(scaler.fit_transform(X), dtype=torch.float32)
```

### Define the Model

```python
class KMeansClustering:
    def __init__(self, num_clusters, max_iters=100, tol=1e-4, random_state=42):
        self.n_clusters = num_clusters
        self.max_iters = max_iters
        self.tol = tol
        self.random_state = random_state
        self.centroids = None
        
    def initialize_centroids(self, X):
       
        if X.shape[0] < self.n_clusters:
            raise ValueError("Number of samples must be greater than number of clusters")
            
        n_samples = X.shape[0]
        if self.random_state is not None:
            torch.manual_seed(self.random_state)
        indices = torch.randint(0, n_samples, (self.n_clusters,))
        self.centroids = X[indices]
        
    def compute_distances(self, X):
        
        return torch.cdist(X, self.centroids)
        
    def assign_clusters(self, X):
       
        distances = self.compute_distances(X)
        return torch.argmin(distances, dim=1)
        
    def update_centroids(self, X, labels):
        
        new_centroids = []
        for i in range(self.n_clusters):
            mask = labels == i
            cluster_points = X[mask]
            if len(cluster_points) > 0:
                new_centroids.append(cluster_points.mean(dim=0))
            else:
                new_centroids.append(self.centroids[i])
        self.centroids = torch.stack(new_centroids)
        
    def fit(self, X):
       
        self.initialize_centroids(X)
        
        for _ in range(self.max_iters):
            labels = self.assign_clusters(X)
            prev_centroids = self.centroids.clone()
            self.update_centroids(X, labels)
            
            if torch.allclose(prev_centroids, self.centroids, atol=self.tol):
                break
                
        return labels
    
    def predict(self, X):
        if self.centroids is None:
            raise ValueError("Model must be fitted before making predictions")
        return self.assign_clusters(X)
    
    def inertia_(self, X, labels):
        return sum(torch.norm(X[labels == i] - self.centroids[i], dim=1).sum().item() 
                  for i in range(self.n_clusters))
```

### Compute within-cluster sum of squares for different k values

```python
def compute_wcss(X, k_values, max_iters=100):
    wcss = []
    for k in k_values:
        kmeans = KMeansClustering(num_clusters=k, max_iters=max_iters)
        labels = kmeans.fit(X)
        wcss.append(kmeans.inertia_(X, labels))
    return wcss
```

### Plot elbow method

```python
k_values = range(1, 11)
wcss = compute_wcss(X_scaled, k_values)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(k_values, wcss, marker='o', linestyle='-')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Within-Cluster Sum of Squares (WCSS)")
plt.title("Elbow Method for Optimal k")
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

### Perform clustering with optimal k (let's use k=3 for this example)

```python
kmeans = KMeansClustering(num_clusters=3, max_iters=100)
labels = kmeans.fit(X_scaled)
```

```python
df['Cluster'] = labels.numpy()
```

### Visualizing the Clusters

```python
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], 
                     c=df['Cluster'], cmap='viridis', s=100)
plt.scatter(kmeans.centroids[:, 1] * scaler.scale_[1] + scaler.mean_[1],
           kmeans.centroids[:, 2] * scaler.scale_[2] + scaler.mean_[2],
           c='red', marker='x', s=200, linewidths=3, label='Centroids')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segmentation using K-Means')
plt.colorbar(scatter, label='Cluster')
plt.legend()
plt.grid(True)
plt.show()
```
